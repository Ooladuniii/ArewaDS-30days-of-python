```python
# Specify the file path
file_path = r'C:\Users\User\Desktop\30-Days-of-Python\data\hacker_news.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
df
```

```python
# Get the number of rows and columns
num_rows, num_columns = df.shape

print(f"Number of Rows: {num_rows}")
print(f"Number of Columns: {num_columns}")
```

```python
# filter the titles which contain python
python_titles = df[df['title'].str.contains('Python', case=False)]
print("Titles containing 'Python':")
python_titles['title']
```

```python
javascript_titles = df[df['title'].str.contains('JavaScript', case=False)]
print("Titles containing 'JavaScript':")
javascript_titles['title']
```
