```python
# Use filter to filter out countries containing 'land'
filtered_countries = list(filter(lambda country: 'land'  in country.lower(), countries))

print(filtered_countries)
```

```python
# Use filter to filter out countries having exactly six characters
filtered_countries = list(filter(lambda country: len(country) != 6, countries))

print(filtered_countries)
```

```python
def to_string(string):
    return str(string)

def get_string_lists(arr):
    return list(map(to_string, arr))

numbers = [1, 2, 3, 4, 5]
result = get_string_lists(numbers)
print(result)
```

```python
from functools import reduce
# Define a custom function for addition
def add(x, y):
    return x + y

# Use reduce to sum all numbers
sum_result = reduce(add, numbers)

print(sum_result)
```

```python
countries = ['Estonia', 'Finland', 'Sweden', 'Denmark', 'Norway', 'Iceland']

# Define a custom function for concatenation
def concatenate_countries(acc, country):
    # If it's the first country, just return the country
    if not acc:
        return country
    # Otherwise, concatenate the current country with the accumulator
    return acc + ', ' + country

# Use reduce to concatenate all countries
sentence = reduce(concatenate_countries, countries)

# Format the final sentence
final_sentence = f"{sentence} are north European countries."

print(final_sentence)
```

```python
def categorize_countries(countries):
    # Initialize empty dictionaries for each category
    land_countries = {'category': 'land', 'countries': []}
    ia_countries = {'category': 'ia', 'countries': []}
    island_countries = {'category': 'island', 'countries': []}
    stan_countries = {'category': 'stan', 'countries': []}

    for country in countries:
        # Check if the country contains the specified patterns
        if 'land' in country.lower():
            land_countries['countries'].append(country)
        if 'ia' in country.lower():
            ia_countries['countries'].append(country)
        if 'island' in country.lower():
            island_countries['countries'].append(country)
        if 'stan' in country.lower():
            stan_countries['countries'].append(country)

    # Create a list to hold all categories
    categorized_countries = [land_countries, ia_countries, island_countries, stan_countries]

    return categorized_countries

# 

result = categorize_countries(countries)

# Print the result
for category in result:
    print(f"{category['category']} countries: {', '.join(category['countries'])}")
```

```python
def count_countries_by_starting_letter(countries):
    starting_letter_count = {}

    for country in countries:
        starting_letter = country[0].upper()
        # Increment the count for the starting letter
        starting_letter_count[starting_letter] = starting_letter_count.get(starting_letter, 0) + 1

    return starting_letter_count


result = count_countries_by_starting_letter(countries)

# Print the result
for letter, count in result.items():
    print(f"Starting letter '{letter}': {count} countries")
```

```python
def get_first_ten_countries(countries):
    return countries[:10]

result = get_first_ten_countries(countries)

# Print the result
print(result)
```

```python
def get_last_ten_countries(countries):
    return countries[-10:]

result = get_last_ten_countries(countries)

# Print the result
print(result)
```
